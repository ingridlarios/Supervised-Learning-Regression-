```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.formula.api as smf 

from os import listdir
listdir(path = '.')

#pandas print options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

#specying file
file = './birthweight_low.xlsx'

#reading file into python
bw_2 = pd.read_excel(io = file)

#outputting the first 10 rows of dataset
bw_2.head(n=10)
```

```python
bw_2.describe().round(4)
```

```python
bw_2.info()
```

```python
#creating columns for missing values, missing value=1 not missing=0

for col in bw_2:
    if bw_2[col].isnull().astype(int).sum() > 0:
        bw_2["m_"+col] = bw_2[col].isnull().astype(int)
        
#checking results of missing values
bw_2[ ["m_meduc", "m_npvis", "m_feduc",] ].sum(axis = 0)
```

```python
#changing all missing values for the mean 

bw_2["meduc"] = bw_2["meduc"].fillna(bw_2["meduc"].mean())

bw_2["npvis"] = bw_2["npvis"].fillna(bw_2["npvis"].mean())

bw_2["feduc"] = bw_2["feduc"].fillna(bw_2["feduc"].mean())
 
```

```python
#verifying that there are no more missing values

#bw_2.isnull().sum(axis = 0)  
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.linear_model 
```

```python
bw_2_data = bw_2.drop (["bwght",
                       "omaps",
                       "fmaps"],
                       axis= 1)

#response variable data
bw_2_target = bw_2.loc[ : , "bwght"]
#log_bw_2_target = bw_2.loc = bw_2.loc[ : , "log_bwght"] #preparing fot later use

# preparing trianing and testing set 
x_train, x_test, y_train, y_test = train_test_split(
        bw_2_data,
        bw_2_target,
        test_size = 0.25,
        random_state = 219)

#shapes of the dataset
print(f""" 
Training Data
----------
X-side: {x_train.shape}
y-side: {y_train.shape}

Testing Data 
-----------
X-side: {x_test.shape}
y side: {y_test.shape}
""")
```

```python
#declaring x variables
x_variables = ["mage", "m_meduc", "meduc", "monpre", "m_npvis", "npvis", "fage", "m_feduc", "feduc", "cigs", "npvis", "drink", "male", "mwhte", "mblck", "moth", "fwhte", "fblck", "foth"]


#looping to make x_variable suitable for statsmodels
for val in x_variables:
     print(f"{val} +")
```

```python
# merging X_train and y_train so that they can be used in statsmodels
bw_2_train = pd.concat([x_train, y_train], axis = 1)


#building ols model
lm_best = smf.ols(formula =  """bwght ~  mage +
                                         m_meduc +
                                         meduc +
                                         m_npvis +
                                         npvis +
                                         fage +
                                         m_feduc +
                                         feduc +
                                         cigs +
                                         drink +
                                         male +
                                         mwhte +
                                         mblck +
                                         moth +
                                         fwhte +
                                         fblck +
                                         foth 
                                         """,
                                data = bw_2_train)
                                         
                                                                                                                                                                           

#fitting model based on the data
results = lm_best.fit()


#analyze the summary output
print(results.summary())
```

```python
#applying scikit-learn

#preparing data 
ols_data = bw_2.loc[ : , x_variables]

#target variable
bw_2_target = bw_2.loc[ : , "bwght"]

#setting train-test split 

x_train_FULL, x_test_FULL, y_train_FULL, y_test_FULL, = train_test_split(
    bw_2_data,        #X
    bw_2_target,      #y
    test_size = 0.25,
    random_state = 219)

#ols the data set for x, y is normal and p-value is included
x_train_OLS, x_test_OLS, y_train_OLS, y_test_OLS = train_test_split(
    ols_data,        #X
    bw_2_target,      #y
    test_size = 0.25,
    random_state = 219)
```

```python
#model object 
lr = LinearRegression() 

#fitting model to training data 
lr_fit = lr.fit(x_train_OLS, y_train_OLS)

#predicting new data
lr_pred = lr_fit.predict(x_test_OLS)

#SCORING results with OLS
print("OLS Training Score: ", lr.score(x_train_OLS, y_train_OLS).round(4))
print("OlS Testing Score: ", lr.score(x_test_OLS, y_test_OLS).round(4))

#data for future use
lr_train_score = lr.score(x_train_OLS, y_train_OLS).round(4) #with r-square value
lr_test_score = lr.score(x_test_OLS, y_test_OLS).round(4)   #with r-sqaure value

#displaying gap between training and test scores
print("OLS Train-Test Gap: ", abs(lr_train_score - lr_test_score).round(4))
lr_test_gap = abs(lr_train_score - lr_test_score).round(4)
```

```python
lr_model_values = zip(bw_2_data[x_variables].columns,
                     lr_fit.coef_.round(decimals = 2))

#placeholdr list for storing of models
lr_model_lst = [("intercept", lr_fit.intercept_.round(decimals = 2))]

#print feature coefficients
for val in lr_model_values:
    lr_model_lst.append(val)
    
#results
for pair in lr_model_lst:
    print(pair)
    
```

```python
#second model (lasso)
lasso_model = sklearn.linear_model.Lasso(alpha = 1.0,
                                         normalize = True)

#Fitting model to train data
lasso_fit = lasso_model.fit(x_train_FULL, y_train_FULL)

#predicting model
lasso_pred = lasso_fit.predict(x_test_FULL)

#scoring results
print("Lasso Training Score: ", lasso_model.score(x_train_FULL, y_train_FULL).round(4))
print("Lasso Testing Score: ", lasso_model.score(x_test_FULL, y_test_FULL).round(4))

lasso_train_score = lasso_model.score(x_train_FULL, y_train_FULL).round(4)
lasso_test_score = lasso_model.score(x_test_FULL, y_test_FULL).round(4)

#gap between test and train scores
print("Lasso Train-Test Gap: ", abs(lasso_train_score - lasso_test_score).round(4))
lasso_test_gap = abs(lasso_train_score - lasso_test_score).round(4)
```

```python
#zipping names to its coef's
lasso_model_values = zip(bw_2_data.columns, lasso_fit.coef_.round(decimals =2))

#setting a placeholder
lasso_model_lst = [("intercept", lasso_fit.intercept_.round(decimals =2))]

#future coef's pairs being printing one by one
for val in lasso_model_values:
    lasso_model_lst.append(val)
    
#checking results
for pair in lasso_model_lst:
    print(pair)
```

```python
#finding coefficients that are equal withint he lasso model
#that are = to 0 and dropping them

#printing coef's
for feature, coefficient in lasso_model_lst:
    
    if coefficient == 0:
        lasso_model_lst.remove((feature, coefficient))
        
#results
for pair in lasso_model_lst:
    print(pair)
```

```python
#initiating an ARD model object to test-train the model 

ard_model = sklearn.linear_model.ARDRegression(normalize = False)

#fitting the model to the training data 
ard_fit = ard_model.fit(x_train_FULL, y_train_FULL)

#predicting the ard model for the new data 
ard_pred = ard_fit.predict(x_test_FULL)

print("Trianing Score: ", ard_model.score(x_train_FULL, y_train_FULL).round(4))
print("Testing Score: ", ard_model.score(x_test_FULL, y_test_FULL).round(4))

#saving scoring data to be able to print later
ard_train_score = ard_model.score(x_train_FULL, y_train_FULL).round(4)
ard_test_score = ard_model.score(x_test_FULL, y_test_FULL).round(4)

#finding and displaying gap for train and test for ard model
print("ARD Train Test Gap: ", abs(ard_train_score - ard_test_score).round(4))
ard_test_gap = abs(ard_train_score - ard_test_score).round(4)
```

```python
#zipping names to its coef's
ard_model_values = zip(bw_2_data.columns, ard_fit.coef_.round(decimals =2))

#setting a placeholder
ard_model_lst = [("intercept", ard_fit.intercept_.round(decimals =2))]

#future coef's pairs being printing one by one
for val in ard_model_values:
    ard_model_lst.append(val)
    
#checking results
for pair in ard_model_lst:
    print(pair)
```

```python
#finding coefficients that are equal withint he lasso model
#that are = to 0 and dropping them

#printing coef's
for feature, coefficient in ard_model_lst:
    
    if coefficient == 0:
        ard_model_lst.remove((feature, coefficient))
        
#results
for pair in ard_model_lst:
    print(pair)
```

```python
#comparing results between all 3 models 

print(f"""
Model     Train Score       Test Score
------    ------------      -----------
OLS         {lr_train_score}          {lr_test_score}
Lasso       {lasso_train_score}          {lasso_test_score}
ARD         {ard_train_score}          {ard_test_score}
""")

#dictionary model 

model_performance = {
    
    "Model Type" : ["OLS", "Lasso", "ARD"],
    
    "Training" : [lr_train_score, lasso_train_score, 
                     ard_train_score],
    
    "Testing" : [lr_test_score, lasso_test_score,
                     ard_test_score],
    
    "Train-Test Gap" : [lr_test_gap, lasso_test_gap, 
                        ard_test_gap],
    
    "Model Size" : [len(lr_model_lst), len(lasso_model_lst),
                    len(ard_model_lst)],
                     
    "Model" : [lr_model_lst, lasso_model_lst, ard_model_lst]}
                     
                    
```
